```python
def xy_to_lonlat(x, y):
    proj_latlon = pyproj.Proj(proj='latlong',datum='WGS84')
    proj_xy = pyproj.Proj(proj="utm", zone=28, datum='WGS84')
    lonlat = pyproj.transform(proj_xy, proj_latlon, x, y)
    return lonlat[1],lonlat[0]
```

```python
xy_to_lonlat(227825.761377,3.193947e+06)
```

```python
#okey transformation, now loop in the dataframe and introduce it into the new 2 columns
```

```python
df['lat_lon'] = df.apply(lambda r: xy_to_lonlat(r['X'],r['Y']),axis=1)
df['Lat'] = df['lat_lon'].apply(lambda x: x[0])
df['Lon'] = df['lat_lon'].apply(lambda x: x[1])
df = df.drop('lat_lon',axis=1)
```

```python
import folium
from folium.plugins import FastMarkerCluster
```

```python
map1 = folium.Map(
      location=[28.6,-17.8],
    tiles='cartodbpositron',
    zoom_start=9,
)
```

```python
df.apply(lambda row:folium.CircleMarker(location=[row["Lat"], row["Lon"]]).add_to(map1), axis=1)
map1
```

```python
map1.save("map1.html")
```

```python

folium_map2 = folium.Map(location=[28.6,-17.8],
                        zoom_start=10,
                        tiles='CartoDB dark_matter')

```
